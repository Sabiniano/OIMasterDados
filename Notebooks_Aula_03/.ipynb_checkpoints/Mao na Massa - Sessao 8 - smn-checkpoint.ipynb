### a) Exiba as 10 primeiras e 10 últimas linhas da tabela

```python
# Importando pandas

import pandas as pd
```

```python
# Colocando o conteúdo do arquivo Ano-2017.xlsx na variável gastos_2017

gastos_2017 = pd.read_excel("Ano-2017.xlsx", sheet_name="Ano-2017")
```

```python
# Exibindo as 10 primeiras linhas

gastos_2017.head(10)
```

```python
# Exibindo as 10 últimas linhas

gastos_2017.tail(10)
```

### b) Quantos registros (linhas) e quantas colunas traz a tabela?

```python
gastos_2017.shape
```

### c) Qual a quantidade de transações (gastos) feitas pelo PT (Partido dos Trabalhadores)?

```python
# Método 1 - Utilizando-se o loc

gastos_2017.loc[gastos_2017["sgPartido"]=="PT"]
```

```python
# Método 2 - Criando o DataFrame gastos_PT e depois aplicando a propriedade index do df e em cima o len

gastos_PT = pd.DataFrame(gastos_2017.loc[gastos_2017["sgPartido"]=="PT"])
len(gastos_PT.index)
```

```python
# Método 3 -  Esta abordagem é menos performática, mas resolve o problema do NaN(Not a Number), pois não conta
# df[df.columns[0]].count()

gastos_PT[gastos_PT.columns[6]].count()
```

### d) Qual o valor médio de transações da base?

```python
# Extraindo o valor médio das transações

media_gastos_2017 = gastos_2017["vlrDocumento"].mean()
media_gastos_2017
```

### e) Qual a quantidade de transações (gastos) acima da média ? (Obs.: Mais tarde vamos mostrar que existe uma baita pegadinha nesse tipo de pergunta).

```python
# Filtrando transações acima da média

qtde_acima = gastos_2017.query("vlrDocumento > @ media_gastos_2017")
qtde_acima[qtde_acima.columns[17]].count()
```

### f) Filtre apenas as transações realizadas por deputados do PSDB do Rio Grande do Sul. Quantas foram as transações?

```python
# Transações realizadas por deputados do PSDB do RS

gastos_2017.loc[(gastos_2017["sgUF"]=="RS") & (gastos_2017["sgPartido"]=="PSDB")]
```

### g) Filtre todas as transações realizadas no período de recesso (suposição: janeiro, fevereiro e dezembro)

```python
# Transações realizadas em janeiro, fevereiro e dezembro

gastos_2017[gastos_2017["numMes"].isin([1, 2, 12])]
```
