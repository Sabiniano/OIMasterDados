### b) Quais são os top 15 municípios com maior percentual de assinantes de tv a cabo? Quais insights chegamos quando avaliamos e comparamos os 2 rankings?

```python
# Calcula o percentual de assinantes de tv a cabo
indsocdem['percentual_tv_cabo'] = indsocdem.apply(lambda row: (row.TV_CABO / row.POP_ESTIMADA) * 100, axis = 1)
# print(indsocdem)
cols = ['CIDADE', 'ESTADO', 'IDHM_Educacao', 'POP_ESTIMADA', 'TV_CABO', 'percentual_tv_cabo']
indsocdem.nlargest(15, 'percentual_tv_cabo')[cols]
```

```python
condicoes = [(indsocdem["IDHM"] < 0.6), (indsocdem["IDHM"] >= 0.6) & (indsocdem["IDHM"] <=0.65), 
                                                  (indsocdem["IDHM"] >= 0.66) & (indsocdem["IDHM"] <=0.75),
                                                   (indsocdem["IDHM"] > 0.75)]
classe = ["1) ate 0.6", "2) entre 0.6 e 0.65", "3) entre 0.66 e 0.75", "4) acima de 0.75"]
indsocdem['Faixas_IDHM'] = np.select(condicoes, classe, default = "Sem Info")
indsocdem.head(3)
```
